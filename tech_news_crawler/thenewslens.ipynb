```python
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 \
            (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',\
            'Accept-Language': 'en-US, en;q=0.5'})

page_num = 5

def get_post(page):
    URL = "https://www.thenewslens.com/category/tech?page={}/".format(page)
    
    webpage = requests.get(URL, headers=HEADERS)
    soup = BeautifulSoup(webpage.content, "html.parser")
    dom = etree.HTML(str(soup))
    date_list = [datetime.strptime(str(s).replace(' ', '').replace('|', ''), '%Y/%m/%d').date() for s in dom.xpath("//span[@class='time']/text()")]
    title_list = [str(s).replace(' ', '') for s in dom.xpath("//h2//a/text()")]
    text_list = [str(s).replace(' ', '') for s in dom.xpath("//p[@class='description']/text()")]
    link_list = [str(s) for s in dom.xpath("//h2//a/@href")]
    df = pd.DataFrame(
    {
        'date': date_list,
        'title': title_list,
        'text': text_list,
        'link':link_list
    })
    return df
```

```python
df = pd.DataFrame(columns=['date', 'title', 'text', 'link'])

for i in range(page_num):
    df = df.append(get_post(i), ignore_index=True)
    sleep(randint(3,15))
```

```python
df.to_csv('thenewslens.csv', encoding='utf-8-sig', columns=['date', 'title', 'text', 'link'])
tmp = pd.read_csv('thenewslens.csv', index_col=0)
tmp
```
