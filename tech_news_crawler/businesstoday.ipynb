```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from lxml import etree
from datetime import datetime
from dateutil.relativedelta import relativedelta
from random import randint
from time import sleep
```

```python
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 \
            (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',\
            'Accept-Language': 'en-US, en;q=0.5'})

page_num = 5

def get_post(page):
    URL = "https://www.businesstoday.com.tw/catalog/183015/list/page/{}/".format(page)
    
    webpage = requests.get(URL, headers=HEADERS)
    soup = BeautifulSoup(webpage.content, "html.parser")
    dom = etree.HTML(str(soup))
    date_list = [datetime.strptime(str(s), '%Y-%m-%d').date() for s in dom.xpath("//p[@class='article__item-date']/text()")]
    title_list = [str(s) for s in dom.xpath("//h4[@class='article__item-title']/text()")]
    text_list = [str(s) for s in dom.xpath("//p[@class='article__item-text']/text()")]
    link_list = [str(s) for s in dom.xpath("//a[@class='article__item']/@href")]
    df = pd.DataFrame(
    {
        'date': date_list,
        'title': title_list,
        'text': text_list,
        'link':link_list
    })
    return df
```

```python
df = pd.DataFrame(columns=['date', 'title', 'text', 'link'])

for i in range(page_num):
    df = df.append(get_post(i), ignore_index=True)
    sleep(randint(3,15))
```

```python
df.to_csv('businesstoday.csv', encoding='utf-8-sig', columns=['date', 'title', 'text', 'link'])
tmp = pd.read_csv('businesstoday.csv', index_col=0)
tmp
```
