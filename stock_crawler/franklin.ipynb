```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.stockq.org/funds/franklin_tw.php"
  
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 \
            (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',\
            'Accept-Language': 'en-US, en;q=0.5'})
  
webpage = requests.get(URL, headers=HEADERS)
soup = BeautifulSoup(webpage.content, "html.parser")
soup
```

```python
table = soup.find('table', {'class': 'fundpricetable'})
table
```

```python
columns = [th.text.replace('\n', '') for th in table.find('tr').find_all('td')]
trs = table.find_all('tr')[1:]
rows = list()
for tr in trs:
    rows.append([td.text for td in tr.find_all('td')])
# ###
df = pd.DataFrame(data=rows, columns=columns)
```

```python
df = df.drop('', axis=1)
df = df[df['基金名稱'].notnull()]
df
```

```python
df.to_csv('franklin.csv', encoding='utf-8-sig')
tmp = pd.read_csv('franklin.csv', index_col=0)
tmp
```
