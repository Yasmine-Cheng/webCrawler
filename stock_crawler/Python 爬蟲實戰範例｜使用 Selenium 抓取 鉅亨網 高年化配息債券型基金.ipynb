# Python 爬蟲實戰範例｜使用 Selenium 抓取鉅亨網高年化配息債券型基金
在本文的主題實作中，將打造基金爬蟲程式，使用 Selenium 套件自動操作瀏覽器，抓取鉅亨網高年化配息債券型基金，會運用到 BeautifulSoup 用來解析網頁內容，並擷取出所需的資料，在使用 Pandas 套件將擷取出來的資料整理好並儲存至 csv 檔案當中

網站文章：https://www.webscrapingpro.tw/cnyes-fund-web-scraping-using-python/

```python
import time
from datetime import datetime
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By

url = 'https://fund.cnyes.com/Fixedincome/search.aspx'

driver = webdriver.Chrome(executable_path='./chromedriver')
driver.get(url=url)

# 計價幣別
# button1 = driver.find_elements(by=By.XPATH, value='//*[@id="ctl00_ContentPlaceHolder1_DD_classCurrent"]')[0]
# button2 = driver.find_elements(by=By.XPATH, value='//*[@id="ctl00_ContentPlaceHolder1_DD_classCurrent"]/option')[11]

# 投資區域（全球市場）
button3 = driver.find_elements(by=By.XPATH, value='//*[@id="ctl00_ContentPlaceHolder1_DropDownList1"]')[0]
button4 = driver.find_elements(by=By.XPATH, value='//*[@id="ctl00_ContentPlaceHolder1_DropDownList1"]/option')[6]

# # 基金組別（高收益債）
# button5 = driver.find_elements(by=By.XPATH, value='//*[@id="ctl00_ContentPlaceHolder1_DropDownList2"]')[0]
# button6 = driver.find_elements(by=By.XPATH, value='//*[@id="ctl00_ContentPlaceHolder1_DropDownList2"]/option')[7]

# # 年化配息率（7%）
# button7 = driver.find_elements(by=By.XPATH, value='//*[@id="aspnetForm"]/div[4]/div[1]/select')[0]
# button8 = driver.find_elements(by=By.XPATH, value='//*[@id="aspnetForm"]/div[4]/div[1]/select[1]/option')[9]

# # 至（10%以上）
# button9 = driver.find_elements(by=By.XPATH, value='//*[@id="aspnetForm"]/div[4]/div[1]/select')[1]
# button10 = driver.find_elements(by=By.XPATH, value='//*[@id="aspnetForm"]/div[4]/div[1]/select[2]/option')[3]

# # 配息頻率
# button11 = driver.find_elements(by=By.XPATH, value='//*[@id="div_type"]')[0]
# button12 = driver.find_elements(by=By.XPATH, value='//*[@id="div_type"]/option')[4]

# 搜尋
button13 = driver.find_elements(by=By.XPATH, value='//*[@id="aspnetForm"]/div[4]/div[3]/button')[0]

# button1.click()
# button2.click()
button3.click()
button4.click()
# button5.click()
time.sleep(5)
# button6.click()
# button7.click()
# button8.click()
# button9.click()
# button10.click()
# time.sleep(5)
# button11.click()
# button12.click()
button13.click()
time.sleep(5)

# 三年績效排序
button14 = driver.find_elements(by=By.XPATH, value='/html/body/div[2]/section[3]/div/div[4]/table[2]/thead/tr/th[3]/select')[0]
button15 = driver.find_elements(by=By.XPATH, value='/html/body/div[2]/section[3]/div/div[4]/table[2]/thead/tr/th[3]/select/option')[6]

button14.click()
button15.click()
time.sleep(5)

html_source = driver.page_source

# driver.close()

soup = BeautifulSoup(html_source, 'lxml')
```

```python
fund_dict = {
    '基金名稱': [],
    '淨值': [],
    '三年績效': [],
    '配息日': [],
    '配息金額': [],
    '年化配息率': [],
    '晨星評級': [],
    '連結': []
}

for tr in soup.select_one('tbody').select('tr'):
    
    morning_star = len(tr.select('td')[5].select('li.on'))
    
    if morning_star >= 3:    
        
        fund_dict['基金名稱'].append(tr.select('td')[0].text.strip())
        fund_dict['淨值'].append(tr.select('td')[1].text.strip().split('\n')[0])
        fund_dict['三年績效'].append(tr.select('td')[2].text.strip())
        fund_dict['配息日'].append(tr.select('td')[3].text.strip().split('\n')[-1].strip())
        fund_dict['配息金額'].append(tr.select('td')[4].text.strip().split('\n')[0].strip())
        fund_dict['年化配息率'].append(tr.select('td')[4].text.strip().split('\n')[-1].strip())
        fund_dict['晨星評級'].append(morning_star)
        fund_dict['連結'].append('https://fund.cnyes.com' + tr.select('td')[0].a['href'].strip().replace(' ', '%20'))
    
fund_dict
```

```python
df = pd.DataFrame(fund_dict)
df.set_index('基金名稱', inplace=True)
df
```

```python
date = datetime.today().strftime("%Y%m%d")
file_name = '{}_全球高收益債基金.csv'.format(date)
df.to_csv(file_name, encoding='utf_8_sig')
```
