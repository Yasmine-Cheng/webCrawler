```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://rili.jin10.com/day/2020-08-17"
  
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 \
            (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',\
            'Accept-Language': 'en-US, en;q=0.5'})
  
webpage = requests.get(URL, headers=HEADERS)
soup = BeautifulSoup(webpage.content, "html.parser")
header_table = soup.find('table', {'class': 'jin-table-header'})
columns = [th.text.replace('\n', '').replace(' ', '') for th in header_table.find('tr').find_all('th')]

body_table = soup.find('table', {'class': 'jin-table-body'})
df = pd.concat(pd.read_html(str(body_table)))
df.columns = columns
df
```
